```python
from model import ImageEncoder
from ultis import predict_image_v2
import torch
```

```python
# Load the pretrained model
model = ImageEncoder()
model.load_state_dict(torch.load("image_encoder_leaf_clip.bin", map_location="cpu"))
model.eval()

# Run inference on test image
img_path = "/kaggle/input/ccmt-subset/CCMT Dataset/Cassava/bacterial blight/bacterial blight1000_.jpg"  # update this path
preds = predict_image_v2(model, img_path, "class_description.json", top_k=5)

# Display the results
print("Top 5 predictions:")
for text, prob in preds:
    print(f"{text}: {prob*100:.2f}%")
```
